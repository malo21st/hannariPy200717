# 文字列で、ちょっと

```python
# 文字列はイテレータ
for char in "abc":
    print(char)
```

```python
# 特殊メソッド　__iter__ を使ってみる
for char in "abc".__iter__():
    print(char)
```

```python
# __iadd__ がなくても、__add__ があれば、累算代入演算をしてくれる。
string = "アルファベット　"
for char in "abc":
    string += char       # string = string + char と等価して処理する
print(string)
```

# 文字列メソッド

```python
"hannari python".title() # upper(), lower(), swapcase()
```

```python
# replace だけじゃない、文字の置き換え maketrans() と translate()
text = "=*=がばい! python =*="
table = str.maketrans('がpば!い', 'はPんりな', '=*') # 処理前文字列， 変換後文字列， 削除対象文字列
text.translate(table)
```

```python
# ファイル名に連番を付ける時に便利　ｚｆｉｌｌ()
for i in range(4):
    print(str(i).zfill(3))
```

```python
set_list__ = { l for l in dir([1,2]) if l[:2] == '__'}
set_tuple__ = { t for t in dir((1,2)) if t[:2] == '__'}
set_dict__ = { d for d in dir({1:'a', 2:'b'}) if d[:2] == '__'}
set_set__ = { s for s in dir({1,2}) if s[:2] == '__'}
```

```python
from venn import venn
%matplotlib inline
%config InlineBackend.figure_format = 'retina'

set_con = {
    "list    （特殊） : 35": set_list__,
    "tuple（特殊） : 31": set_tuple__,
    "dict  （特殊） : 29": set_dict__,
    "set   （特殊） : 38": set_set__
}
fig = venn(set_con, cmap="cool")
fig
```

```python
set1 = set_list__ & set_tuple__ & set_dict__ & set_set__
print(len(set1), set1)
```

```python
set2 = set_list__ & set_tuple__ & set_dict__ - set_set__
print(len(set2), set2)
```
